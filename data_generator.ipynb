```python
def create_spark_session():
    import pyspark
    import os
    from pyspark.sql import SparkSession
    spark=SparkSession.builder.appName('Dataset').getOrCreate()
    os.environ['PYSPARK_PYTHON'] = "C:/Users/Administrator/AppData/Local/Programs/Python/Python311/python.exe"
    os.environ['PYSPARK_DRIVER_PYTHON'] = "C:/Users/Administrator/AppData/Local/Programs/Python/Python311/python.exe"
    
    return spark
```

```python
def generate_data():
    acNumbers = [ num for num in range(670001, 670040) ]
    tranType = ['C', 'D']
    
    count = 100
    for x in range(0, 100):
        data = {}
        readings = []
        for iter in range(0, 10):
            for i in range(0, 20):
                data['Acc No'] = acNumbers[random.randint(0,38)]
                data['Transaction'] = tranType[random.randint(0,1)]
                data['Time Stamp'] = str(datetime.now())
                data['Amount'] = random.randint(1,1000000)
                readings.append(data.copy())
            time.sleep(1)
        fname = 'trans'+str(count)+'.json'
        files = open('./data/'+fname, 'w')
        files.write(json.dumps(readings))
        files.close()
        count += 1
```

```python
# import_functions()
create_spark_session()
generate_data()
```
