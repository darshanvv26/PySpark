```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .appName("RDD Ex1") \
    .master("local[1]") \
    .getOrCreate()

import os, sys
from pyspark.sql import SparkSession

# Force Spark to use the same Python
os.environ["PYSPARK_PYTHON"] = sys.executable
os.environ["PYSPARK_DRIVER_PYTHON"] = sys.executable
```

```python
some_text = ['Hope all is well', 'ABD and ADS classes are not boring', 'ABD and ADS are Architecture of BigData and Algorithms and Data Structures', 'Hope all going good']
parRdd = spark.sparkContext.parallelize(some_text)
```
