```python
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder.appName('DateTime').getOrCreate()
```

```python
# withColums() is used to add new column to the existing data frame
df = spark.range(5).withColumn('Today', current_date()).withColumn('New', current_timestamp())
```

```python
df.show()
```

```python
df = df.withColumn('New Date', date_add('Today', 5))
df = df.withColumn('Prev Date', date_sub('Today', 5))
df.show()
```

```python
df.select(abs(datediff('Prev Date', 'New Date'))).show()
```

```python
df = df.withColumn('String Date', lit('21-10-2025'))
df.show()
```

```python
df.printSchema()
```

```python
df = df.withColumn('Proper Date', to_date('String Date', 'dd-MM-yyyy'))
df.show()
```

```python
df = df.withColumn('Formatted Date', date_format(to_date('String Date', 'dd-MM-yyyy'), 'dd-MM-yyyy'))
df.show()
```

```python
df.printSchema()
```

```python
df.select(quarter(to_date(lit('2025-10-16')))).show()
```

```python
df.select(year(to_date(lit('2025-10-16')))).show()
```

```python
df.select(month(to_date(lit('2025-10-16')))).show()
```

```python
df.select(weekofyear(to_date(lit('2025-10-16')))).show()
```

```python
df.select(dayofweek(to_date(lit('2025-10-16')))).show()
```
