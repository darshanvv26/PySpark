```python
spark = SparkSession.builder.appName('Examplessss').getOrCreate()
```

```python
some_text = ['do not trouble the trouble', 'if you trouble the trouble', 'trouble troubles you']
parRdd = spark.sparkContext.parallelize(some_text)
```

```python
flatRdd = parRdd.flatMap(lambda d:d.split(' '))
mapRdd = flatRdd.map(lambda d:(d,1))
frequencyRdd = mapRdd.reduceByKey(lambda a,b:a+b)
frequencyRdd
```
